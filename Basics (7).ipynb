```python
duplicate_apps = []
unique_apps = []
for each in google_data:
    app_name = each[0]
    if app_name in unique_apps:
        duplicate_apps.append(app_name)
    unique_apps.append(app_name)
print(len(unique_apps))
print('\n')
print(len(duplicate_apps))
print('\n')
print(duplicate_apps[1:15])
```

```python
app_names = []
for each in google_data:
    name =each[0]
    app_names.append(name)
```

```python
print(app_names[:10])
```

```python
app_freq = {}
for each in app_names:
    if each in app_freq:
        app_freq[each] +=1
    else:
        app_freq[each] = 1
```

```python
for each in app_freq:
    if app_freq[each] >=2:
        print(each)
        print('\n')
        
        
```

#As shown above the Google data has 1811 duplicated data that need to be removed.

```python
reviews_max = {}

for app in google_main:
    name = app[0]
    n_reviews = float(app[3])
    
    if name in reviews_max and reviews_max[name] < n_reviews:
        reviews_max[name] = n_reviews
        
    elif name not in reviews_max:
        reviews_max[name] = n_reviews
 
```
